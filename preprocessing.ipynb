```python
link="https://github.com/CAP-CoyunturasCriticas/BookMaps2011/raw/refs/heads/main/Elecciones%202011_Distrital.xlsx"

import pandas as pd

e2011p=pd.read_excel(link,dtype={'UBIGEO': object})
e20112=pd.read_excel(link,1,dtype={'UBIGEO': object})
```

```python
e20112.columns
```

```python
e2011p=e2011p.loc[:,['DEPARTAMENTO', 'PROVINCIA', 'DISTRITO', 'UBIGEO','GANA PERU','VOTOS VÁLIDOS']]
e20112=e20112.loc[:,['DEPARTAMENTO', 'PROVINCIA', 'DISTRITO', 'UBIGEO','GANA PERU','VOTOS VALIDOS']]
```

```python
e2011p['Ollanta1']=e2011p['GANA PERU']/e2011p['VOTOS VÁLIDOS']
e20112['Ollanta2']=e20112['GANA PERU']/e20112['VOTOS VALIDOS']
```

```python
e2011p=e2011p[~e2011p.UBIGEO.str.startswith('9')]
e20112=e20112[~e20112.UBIGEO.str.startswith('9')]
```

```python
e2011p.info()
```

```python
e20112.info()
```

```python
e2011=e2011p.merge(e20112,on=['DEPARTAMENTO', 'PROVINCIA', 'DISTRITO', 'UBIGEO'])
e2011=e2011.loc[:,['DEPARTAMENTO', 'PROVINCIA', 'DISTRITO', 'UBIGEO','Ollanta1','Ollanta2']]


from unidecode import unidecode

e2011['DISTRITO']=e2011.DISTRITO.apply(unidecode)
e2011['PROVINCIA']=e2011.PROVINCIA.apply(unidecode)
```
